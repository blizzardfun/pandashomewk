```python
# Dependencies and Setup
import pandas as pd
#import numpy as np

# Raw data file
file_to_load = "Resources/purchase_data.csv"

# Read purchasing file and store into pandas data frame
purch_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
total_players = len(purch_data["SN"].unique())                          #find total unique SN
total_player_df=pd.DataFrame([{"Total Players" :total_players}])   #put it in a dataframe and print
total_player_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_item_count =len(purch_data["Item ID"].unique())       #count of unique item ids
ave_purch_price = purch_data["Price"].mean()                 # average price of a purchase

total_revenue = purch_data["Price"].sum()                   #total purchases
purch_count = purch_data["Price"].count()                  #how many purchases
```

```python
#results into DataFrame and format
summary_data=pd.DataFrame([{"Number of Unique Items":unique_item_count,"Average Purchase Price":ave_purch_price,"Total Purchases": purch_count,"Total Revenue":total_revenue}])
summary_data=summary_data[["Number of Unique Items", "Average Purchase Price","Total Purchases","Total Revenue" ]]
summary_data["Average Purchase Price"]=summary_data["Average Purchase Price"].map("${:,.2f}".format)
summary_data["Total Revenue"]=summary_data["Total Revenue"].map("${:,.2f}".format)
summary_data
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_player_df=purch_data[["Gender","SN"]]                 #remove extranious data
gender_player_df =gender_player_df.drop_duplicates()           # remove duplicates / get unique screennames
gender_grp=gender_player_df.groupby(["Gender"])         # group by gender
gender_player_count=gender_grp["SN"].count()             #count the Screen Names

gender_count_df=pd.DataFrame(gender_player_count)                        #change to a dataframe so I can add a column
gender_count_df["Percent of Total"]=(gender_count_df.iloc[:,0]/total_players)*100                #add the percentage column
new_gender_count_df=gender_count_df.rename(columns={"SN":"Player Count"})        # make column names better
new_gender_count_df["Percent of Total"]=new_gender_count_df["Percent of Total"].map("{:.2f}%".format)
new_gender_count_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_grp = purch_data.groupby(["Gender"])              # group by gender
gender_purch_count=gender_grp["Purchase ID"].count()         #find count,average purchase price and total purchases by group
gender_purch_ave=gender_grp["Price"].mean()              
gender_purch_tot=gender_grp["Price"].sum()               

#--------------------------------------
gender_player_grp=purch_data.groupby(["Gender","SN"])                 #grouped by gender and player
gender_player_purch =gender_player_grp["Price"].sum()                 # Purchase Total per Person by Gender

new_gender_player_purch=gender_player_purch.reset_index(drop = False)  #remove indexes from double groupby
new_gender_purch_grp =new_gender_player_purch.groupby(["Gender"])         #re-add groupby gender
ave_gender_purch=new_gender_purch_grp["Price"].mean()          #Average Purchase Total per Person by Gender
#ave_gender_purch
```

```python
#results to a dataframe and format
gender_summary_df=pd.DataFrame({"Purchase Count":gender_purch_count,"Average Purchase Price":gender_purch_ave,"Total Purchase Value":gender_purch_tot,"Ave Purchase Total Per Person":ave_gender_purch})
gender_summary_df["Average Purchase Price"]=gender_summary_df["Average Purchase Price"].map("${:,.2f}".format)
gender_summary_df["Total Purchase Value"]=gender_summary_df["Total Purchase Value"].map("${:,.2f}".format)
gender_summary_df["Ave Purchase Total Per Person"]=gender_summary_df["Ave Purchase Total Per Person"].map("${:,.2f}".format)
gender_summary_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purch_data["age group"]=pd.cut(purch_data["Age"],age_bins,labels=group_names)

age_player_plus_df=purch_data.loc[:,["age group","SN"]]            #reduce columns to age group and SN
age_player_df=age_player_plus_df.drop_duplicates()                  #remove duplicates (from multi purchases)
age_grp=age_player_df.groupby(["age group"])                      #make groups

age_player_count=age_grp["SN"].count()                            #count unique SN
age_player_count_df=pd.DataFrame(age_player_count)                 #put in dataframe
age_player_count_df["Percentage of Total"]=(age_player_count_df.iloc[:,0]/total_players)*100    # calc percentages
summary_age_df=age_player_count_df.rename(columns={"SN":"Player Count"})                          #format
summary_age_df["Percentage of Total"]=summary_age_df["Percentage of Total"].map("{:,.2f}%".format)

summary_age_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_purch_grp=purch_data.groupby(["age group"])                #find count,average price and total price by age
age_purch_count=age_purch_grp["SN"].count()
age_purch_ave=age_purch_grp["Price"].mean()
age_purch_tot=age_purch_grp["Price"].sum()
#--------------------------------------
age_player_grp=purch_data.groupby(["age group","SN"])                 #grouped by age and player
age_player_purch =age_player_grp["Price"].sum()                 # Purchase Total per Person by age

new_age_player_purch=age_player_purch.reset_index(drop = False)  #remove indexes from double groupby
new_age_purch_grp =new_age_player_purch.groupby(["age group"])         #re-add groupby age 
ave_age_purch=new_age_purch_grp["Price"].mean()          #Average (Purchase Total per Person) by age
#ave_age_purch
```

```python
#results to a dataframe and format
age_summary_df=pd.DataFrame({"Purchase Count":age_purch_count,"Average Purchase Price":age_purch_ave,"Total Purchase Value":age_purch_tot,"Average Purchase Total Per Person":ave_age_purch})
age_summary_df["Average Purchase Price"]=age_summary_df["Average Purchase Price"].map("${:,.2f}".format)
age_summary_df["Total Purchase Value"]=age_summary_df["Total Purchase Value"].map("${:,.2f}".format)
age_summary_df["Average Purchase Total Per Person"]=age_summary_df["Average Purchase Total Per Person"].map("${:,.2f}".format)
age_summary_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
player_grp=purch_data.groupby(["SN"])                 #grouped by player
player_purch_count=player_grp["Price"].count()           # count purchases per player

player_purch_ave=player_grp["Price"].mean()                #average purchase price
player_purch_total =player_grp["Price"].sum()                 # Purchase Total per player
                   #dataframe summary of results
player_purch_df= pd.DataFrame({"Purchase Count":player_purch_count,"Average Purchase Price":player_purch_ave,"Total Purchase Value":player_purch_total})
player_purch_df=player_purch_df.sort_values("Total Purchase Value",ascending=False)         #sort by total purchase value
player_purch_df["Average Purchase Price"]=player_purch_df["Average Purchase Price"].map("${:,.2f}".format)   #add formatting
player_purch_df["Total Purchase Value"]=player_purch_df["Total Purchase Value"].map("${:,.2f}".format)
player_purch_df.head(10)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_df=purch_data[["Item ID","Item Name","Price"]]
item_df=item_df.rename(columns={"Price":"Item Price"})        #rename column

item_grp=item_df.groupby(["Item ID","Item Name"])          #group by the item (id and name)
item_purch_count =item_grp["Item ID"].count()             #count rows per item
item_purch_total=item_grp["Item Price"].sum()              #sum of price per item
item_df=item_df.drop_duplicates()                         #drop duplicates so one row per item

item_df_2=pd.DataFrame({"Purchase Count": item_purch_count,"Total Purchase Value":item_purch_total})  
                #new item stats
merge_item_df=pd.merge(item_df, item_df_2,how ="left",on = ["Item ID","Item Name"])       
            #merge with original item data

merge_item_df=merge_item_df[["Item ID","Item Name","Purchase Count","Item Price","Total Purchase Value"]] #rearrange summary data
sort_item_df=merge_item_df.sort_values("Purchase Count",ascending = False)                        #sort descending
sort_item_df["Item Price"]=sort_item_df["Item Price"].map("${:,.2f}".format)                            #format data
sort_item_df["Total Purchase Value"]=sort_item_df["Total Purchase Value"].map("${:,.2f}".format)
#sort_item_df["Item Name"]=sort_item_df["Item Name"].astype(str).map("{:50}".format)      #??can't left justify item name
#print(" {:^70}".format(sort_item_df.iloc[0,1]))             
sort_item_df.head()
```

* Sort the above table by total purchase value in descending order

```python
new_item_df=merge_item_df.sort_values("Total Purchase Value",ascending = False)
new_item_df["Item Price"]=new_item_df["Item Price"].map("${:,.2f}".format)                            #format data
new_item_df["Total Purchase Value"]=new_item_df["Total Purchase Value"].map("${:,.2f}".format)
new_item_df.head()
```
